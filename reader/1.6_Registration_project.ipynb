<b>TODO Yousra: Check the below explanation makes sense and rewrite if necessary </b>

<font color='blue'> To estimate the registration error, I would use corresponding point pairs that were not used to compute the affine transformation. These points act as independent validation points. After applying the affine transformation to the moving image points, I would compare their transformed positions with the corresponding points in the fixed/reference image. The registration error can then be calculated as the Euclidean distance between each transformed point and its true corresponding point. The mean or root mean square error over all validation points gives an estimate of the registration accuracy. </font>
 
<font color='blue'> The same points that were used to compute the affine transformation should not be used to estimate the registration error, because this would give an overly optimistic error / a very high bias. The transformation is fitted to those points, so the error on them may be small even if the registration is not accurate for other parts of the image. </font>

<font color='blue'> The number of corresponding point pairs influences the registration error. A 2D affine transformation needs at least three pairs because of the three degrees of freedom, but using only three is sensitive to noise. More accurate and well-distributed points usually improve robustness through least-squares fitting, while inaccurate or poorly placed points can increase the error. </font>
