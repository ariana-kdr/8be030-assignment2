5. Your answer should include a description of the feature, the code to extract it, and a demonstration that shows it is different from the features you already have.
   
The new feature was an edge detection operator, specifically a Sobel filter (`scipy.ndimage.sobel()`). Besides Sobel, two other edge detection operators were considered: the Prewett operator, and the Canny operator. However, Prewett uses a uniform averaging technique which makes it less accurate than Sobel. While Canny is a more advanced operator, it requires threshold values for pixel brightnesses to be able to classify strong and weak edges. Since this information was not readily available, Sobel was ultimately chosen. The choice of a Sobel filter was also validated by literature.
https://www.sciencedirect.com/science/article/pii/S1877050922004768
https://link.springer.com/article/10.1007/s10462-025-11151-8

The Sobel filter detects how rapidly pixel intensities change across neighbouring pixels. It scans the image pixels across both axes to detect horizontal and vertical edges by computing gradients. Both these results were then combined to find overall edge strengths. Higher values in the feature vector means there is a boundary across which intensity values are changing, whilst lower values mean the area is homogeneous. Two new columns were added to the features matrix: the Sobel filter results applied to the T1 image, and to the T2 image. Both T1 and T2 were used given that the boundary features appear visually different in both. 

Across the comparisons, the Sobel feature showed no strong relationship with the other existing features. For any given value of intensity, Gaussian smoothing, local variance, or T1-T2 difference, it could be seen that the Sobel values for the tissue (white matter, grey matter and CSF) remained low (0-300) with significant scatter. The background and skull showed scatter across the entire Sobel range (0-1000). This compression of the tissue values into a smaller range is not something that appears in the other features being used, and so it provides useful information about distinguishing between the background and the tissue.

```python
# OVERALL CHANGE IN DICE SCORES WITH AND WITHOUT EDGE DETECTION 
```

```python
# TISSUE SPECIFIC CHANGE IN DICE SCORES

def dice_per_class(y_true, y_pred, class_names={0:'Background', 1:'White matter', 2:'Gray matter', 3:'CSF'}):
    scores = {}
    for c, name in class_names.items():
        true_c = (y_true == c)
        pred_c = (y_pred == c)
        intersection = np.sum(true_c & pred_c)
        dice = 2 * intersection / (np.sum(true_c) + np.sum(pred_c) + 1e-8)
        scores[name] = dice
    return scores

model_no_edge = RandomForestClassifier(n_estimators=100, min_samples_leaf=2, random_state=0, n_jobs=-1)
model_edge    = RandomForestClassifier(n_estimators=100, min_samples_leaf=2, random_state=0, n_jobs=-1)

X_train, y_train, _ = util.create_dataset(1, 1, 'tissue', include_edge=True)
X_test, y_test, _   = util.create_dataset(2, 1, 'tissue', include_edge=True)
y_train, y_test = y_train.ravel(), y_test.ravel()

cols_no_edge = [0,1,2,3,4,5,6]
cols_edge    = list(range(X_train.shape[1]))

model_no_edge.fit(X_train[:,cols_no_edge], y_train)
model_edge.fit(X_train[:,cols_edge], y_train)

pred_no_edge = model_no_edge.predict(X_test[:,cols_no_edge])
pred_edge    = model_edge.predict(X_test[:,cols_edge])

dice_no_edge = dice_per_class(y_test, pred_no_edge)
dice_edge    = dice_per_class(y_test, pred_edge)

print('Tissue       | No edge | With edge | Difference')
for name in dice_no_edge:
    diff = dice_edge[name] - dice_no_edge[name]
    print(f'{name:13s} | {dice_no_edge[name]:.4f}  | {dice_edge[name]:.4f}    | {diff:+.4f}')
```

```python
# CONFUSION MATRIX
# Might be nice to normalise this to be in terms of percentages 

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

fig, axes = plt.subplots(1, 2, figsize=(10, 4))
labels = ['Background', 'WM', 'GM', 'CSF']

cm_no_edge = confusion_matrix(y_test, pred_no_edge)
cm_edge    = confusion_matrix(y_test, pred_edge)

ConfusionMatrixDisplay(cm_no_edge, display_labels=labels).plot(ax=axes[0])
axes[0].set_title('Without edge features')

ConfusionMatrixDisplay(cm_edge, display_labels=labels).plot(ax=axes[1])
axes[1].set_title('With edge features')

plt.tight_layout()
plt.show()
```

```python
# DIFFERENCE MAP

util.show_segmentation(y_test, pred_no_edge, title='Without edge features')
util.show_segmentation(y_test, pred_edge, title='With edge features')

# Difference map: where do the two predictions disagree?
# would be nice to overlay this on the actual ground truth mask
diff_map = (pred_no_edge != pred_edge).reshape(240, 240)
plt.figure(figsize=(5,5))
plt.imshow(diff_map, cmap='Reds')
plt.title('Pixels where edge features changed the prediction')
plt.axis('off')
plt.show()
```

```python
# LEARNING CURVE

def learning_curve_dice(X_train, y_train, X_test, y_test, cols, sample_sizes, model_fn):
    dice_scores = []
    for n in sample_sizes:
        idx = np.random.choice(len(y_train), size=min(n, len(y_train)), replace=False)
        model = model_fn()
        model.fit(X_train[idx][:, cols], y_train[idx])
        y_pred = model.predict(X_test[:, cols])
        dice = util.dice_multiclass(y_test.reshape(-1,1), y_pred.reshape(-1,1))
        dice_scores.append(dice)
    return dice_scores

sample_sizes = [500, 1000, 2500, 5000, 10000, 20000]
model_fn = lambda: RandomForestClassifier(n_estimators=100, min_samples_leaf=2, random_state=0, n_jobs=-1)

dice_curve_no_edge = learning_curve_dice(X_train, y_train, X_test, y_test, cols_no_edge, sample_sizes, model_fn)
dice_curve_edge    = learning_curve_dice(X_train, y_train, X_test, y_test, cols_edge, sample_sizes, model_fn)

plt.figure(figsize=(6,4))
plt.plot(sample_sizes, dice_curve_no_edge, marker='o', label='Without edge features')
plt.plot(sample_sizes, dice_curve_edge, marker='o', label='With edge features')
plt.xlabel('Number of training samples')
plt.ylabel('Multiclass Dice score')
plt.title('Learning curve: effect of edge features')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
