# Project: Image segmentation

## Wat doen we eigenlijk?

We willen een methode maken die **voor elke pixel in een hersenbeeld** voorspelt tot welke klasse die pixel hoort:

- `0` = background
- `1` = white matter
- `2` = gray matter
- `3` = CSF / cerebrospinal fluid

## Research question

**Does adding edge-based features improve tissue segmentation compared with using intensity and smoothing features only?**

In normale woorden: we testen of informatie over randen/borders tussen weefsels helpt om white matter, gray matter en CSF beter te onderscheiden.

## Hypothesis

Edge features zullen de Dice score verbeteren, omdat tissue boundaries vaak sterke intensiteitsveranderingen hebben. De verbetering kan wel beperkt zijn als tissues qua intensiteit sterk overlappen.

# 0. Imports

```python
%matplotlib inline
import sys
sys.path.append("../code")

import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.decomposition import PCA
import segmentation_util as util
import segmentation as seg
import segmentation_project as project

np.random.seed(0)
```

# 1. Dataset maken

De functie `create_dataset(image_number, slice_number, task)` doet twee dingen:

1. `extract_features(...)` maakt per pixel een rij met features.
2. `create_labels(...)` maakt per pixel het correcte label.

Voor dit project gebruiken we `task='tissue'`, want we willen white matter vs gray matter vs CSF vs background.

```python
# Kies eerst één train subject en één test subject.
# Je kunt dit later aanpassen voor extra experimenten.
train_subject = 1
test_subject = 2
slice_number = 1
task = 'tissue'

X_train, y_train, feature_labels = util.create_dataset(train_subject, slice_number, task)
X_test, y_test, _ = util.create_dataset(test_subject, slice_number, task)

y_train = y_train.ravel()
y_test = y_test.ravel()

print('Training data shape:', X_train.shape)
print('Test data shape:', X_test.shape)
print('Feature labels:')
for i, f in enumerate(feature_labels):
    print(i, f)
print('Classes in y_train:', np.unique(y_train))
```

# 2. Wat zijn onze features?

De eerste twee features waren al gegeven:

- T1 intensity
- T2 intensity

Daar hebben we extra features aan toegevoegd:

- Gaussian smoothing van T1 en T2: lokale/contextuele intensiteit
- T1 - T2 verschil: verschil tussen beide MRI-kanalen
- Sobel edge strength: randsterkte, dus waar intensiteit sterk verandert

Waarom edge strength? Omdat boundaries tussen hersenweefsels vaak intensiteitsranden zijn. Dit is onze eigen extra feature voor de research question.

```python
# Scatterplots: hiermee kijken we of features de tissue classes een beetje scheiden.
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

util.scatter_data(X_train, y_train, feature0=0, feature1=1, ax=axes[0])
axes[0].set_xlabel(feature_labels[0])
axes[0].set_ylabel(feature_labels[1])
axes[0].set_title('T1 vs T2')

util.scatter_data(X_train, y_train, feature0=2, feature1=3, ax=axes[1])
axes[1].set_xlabel(feature_labels[2])
axes[1].set_ylabel(feature_labels[3])
axes[1].set_title('Gaussian features')

util.scatter_data(X_train, y_train, feature0=7, feature1=8, ax=axes[2])
axes[2].set_xlabel(feature_labels[7])
axes[2].set_ylabel(feature_labels[8])
axes[2].set_title('Edge features')

plt.tight_layout()
plt.show()
```

```python
# PCA plot: samenvatting van alle features in 2 dimensies.
# Dit helpt om te zien of de klassen globaal gescheiden zijn.
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_train)
X_pca = PCA(n_components=2).fit_transform(X_scaled)

idx = np.random.choice(len(y_train), size=min(3000, len(y_train)), replace=False)
plt.figure(figsize=(6, 5))
plt.scatter(X_pca[idx, 0], X_pca[idx, 1], c=y_train[idx], s=3)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA plot of features')
plt.colorbar(label='class')
plt.show()
```

## interpretatie

Als de kleuren/classes in de scatterplots veel overlappen, betekent dit dat sommige tissues op elkaar lijken in feature space. Dan wordt segmentatie moeilijker.

Als edge features extra scheiding geven, verwachten we dat een classifier met edge features beter presteert dan een classifier zonder edge features.

# 3. Evaluatiefuncties

We gebruiken:

- **classification error**: percentage pixels dat fout is voorspeld
- **multiclass Dice**: overlap tussen voorspelde mask en ground truth, gemiddeld over de classes

Dice is belangrijker voor segmentatie, omdat we echt de overlap van masks willen beoordelen.

```python
def evaluate_model(model, X_train, y_train, X_test, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    err = util.classification_error(y_test, y_pred)
    dice = util.dice_multiclass(y_test.reshape(-1,1), y_pred.reshape(-1,1))
    return err, dice, y_pred


def show_segmentation(y_true, y_pred, title='Segmentation result'):
    true_mask = y_true.reshape(240, 240)
    pred_mask = y_pred.reshape(240, 240)

    fig, axes = plt.subplots(1, 2, figsize=(10, 4))
    axes[0].imshow(true_mask)
    axes[0].set_title('True mask')
    axes[0].axis('off')
    axes[1].imshow(pred_mask)
    axes[1].set_title(title)
    axes[1].axis('off')
    plt.tight_layout()
    plt.show()
```

# 4. Model selection

We vergelijken feature-combinaties:

1. Alleen intensiteit: `T1 + T2`
2. Intensiteit + smoothing
3. Intensiteit + smoothing + verschil
4. Alle features inclusief edge

We gebruiken eerst kNN omdat dit aansluit bij de oefeningen. Daarna testen we Random Forest als eigen methode.

```python
feature_sets = {
    'T1 + T2 only': [0, 1],
    'T1/T2 + Gaussian': [0, 1, 2, 3, 4, 5],
    'T1/T2 + Gaussian + difference': [0, 1, 2, 3, 4, 5, 6],
    'All features including edge': list(range(X_train.shape[1]))
}

models = {
    'kNN k=3': make_pipeline(StandardScaler(), KNeighborsClassifier(n_neighbors=3)),
    'kNN k=7': make_pipeline(StandardScaler(), KNeighborsClassifier(n_neighbors=7)),
    'Random Forest': RandomForestClassifier(n_estimators=100, min_samples_leaf=2, random_state=0, n_jobs=-1)
}

results = []
predictions = {}

# Subsample training pixels so the notebook runs faster.
N = min(10000, len(y_train))
idx_train = np.random.choice(len(y_train), size=N, replace=False)

for fs_name, cols in feature_sets.items():
    for model_name, model in models.items():
        err, dice, y_pred = evaluate_model(
            model,
            X_train[idx_train][:, cols], y_train[idx_train],
            X_test[:, cols], y_test
        )
        results.append((fs_name, model_name, err, dice))
        predictions[(fs_name, model_name)] = y_pred

# Print results sorted by Dice score
results_sorted = sorted(results, key=lambda x: x[3], reverse=True)
print('Feature set | Model | Error | Dice')
for fs_name, model_name, err, dice in results_sorted:
    print(f'{fs_name:35s} | {model_name:15s} | {err:.4f} | {dice:.4f}')

best_fs, best_model, best_err, best_dice = results_sorted[0]
print('Best choice:')
print('Features:', best_fs)
print('Model:', best_model)
print('Error:', best_err)
print('Dice:', best_dice)
```

```python
# Visualize the best result
best_pred = predictions[(best_fs, best_model)]
show_segmentation(y_test, best_pred, title=f'Prediction: {best_model}, {best_fs}')
```

# 5. Research question experiment

Research question:

**Does adding edge-based features improve tissue segmentation compared with using intensity and smoothing features only?**

Daarvoor vergelijken we specifiek:

- zonder edge features
- met edge features

We houden het model hetzelfde, zodat het verschil vooral door de features komt.

```python
# Compare without edge vs with edge using the same classifier
without_edge_cols = [0, 1, 2, 3, 4, 5, 6]
with_edge_cols = list(range(X_train.shape[1]))

model_without_edge = make_pipeline(StandardScaler(), KNeighborsClassifier(n_neighbors=7))
model_with_edge = make_pipeline(StandardScaler(), KNeighborsClassifier(n_neighbors=7))

err_no_edge, dice_no_edge, pred_no_edge = evaluate_model(
    model_without_edge,
    X_train[idx_train][:, without_edge_cols], y_train[idx_train],
    X_test[:, without_edge_cols], y_test
)

err_edge, dice_edge, pred_edge = evaluate_model(
    model_with_edge,
    X_train[idx_train][:, with_edge_cols], y_train[idx_train],
    X_test[:, with_edge_cols], y_test
)

print('Without edge features')
print('Error:', err_no_edge)
print('Dice:', dice_no_edge)

print('With edge features')
print('Error:', err_edge)
print('Dice:', dice_edge)

print('Difference in Dice:', dice_edge - dice_no_edge)
```

```python
show_segmentation(y_test, pred_no_edge, title='Prediction without edge features')
show_segmentation(y_test, pred_edge, title='Prediction with edge features')
```

# 6. Generalisatie naar ander subject

Nu herhalen we eigenlijk hetzelfde experiment met een ander test subject. Dit laat zien of onze conclusie stabiel blijft.

```python
def run_edge_experiment(train_subject=1, test_subject=3, slice_number=1):
    X_tr, y_tr, _ = util.create_dataset(train_subject, slice_number, 'tissue')
    X_te, y_te, _ = util.create_dataset(test_subject, slice_number, 'tissue')
    y_tr = y_tr.ravel()
    y_te = y_te.ravel()

    idx = np.random.choice(len(y_tr), size=min(10000, len(y_tr)), replace=False)

    model_no_edge = make_pipeline(StandardScaler(), KNeighborsClassifier(n_neighbors=7))
    model_edge = make_pipeline(StandardScaler(), KNeighborsClassifier(n_neighbors=7))

    err1, dice1, pred1 = evaluate_model(
        model_no_edge,
        X_tr[idx][:, without_edge_cols], y_tr[idx],
        X_te[:, without_edge_cols], y_te
    )

    err2, dice2, pred2 = evaluate_model(
        model_edge,
        X_tr[idx][:, with_edge_cols], y_tr[idx],
        X_te[:, with_edge_cols], y_te
    )

    print(f'Train subject {train_subject}, test subject {test_subject}, slice {slice_number}')
    print(f'Without edge: error={err1:.4f}, dice={dice1:.4f}')
    print(f'With edge:    error={err2:.4f}, dice={dice2:.4f}')
    print(f'Dice difference: {dice2-dice1:.4f}')

    return dice1, dice2, pred1, pred2, y_te

dice1_s3, dice2_s3, pred1_s3, pred2_s3, y_test_s3 = run_edge_experiment(train_subject=1, test_subject=3, slice_number=1)
```

# 7. Combining methods:

In `segmentation_project.py` is `segmentation_mymethod` ingevuld met een Random Forest.

Waarom Random Forest?

- Het kan non-lineaire combinaties van features leren.
- Het werkt voor multiclass segmentation.
- Het is minder gevoelig voor schaling dan kNN.
- Het kan intensiteit, smoothing en edge features tegelijk gebruiken.

```python
# Test project.segmentation_mymethod on the tissue task.
# We bouwen hiervoor train_data_matrix en train_labels_matrix zoals segmentation_demo dat ook doet.
train_subjects = [1, 2, 3, 4]
test_subject = 5
slice_number = 1

data_list = []
label_list = []
for sub in train_subjects:
    X_sub, y_sub, _ = util.create_dataset(sub, slice_number, 'tissue')
    data_list.append(X_sub)
    label_list.append(y_sub.ravel())

train_data_matrix = np.stack(data_list, axis=2)
train_labels_matrix = np.stack(label_list, axis=1)

X_te, y_te, _ = util.create_dataset(test_subject, slice_number, 'tissue')
y_te = y_te.ravel()

pred_my = project.segmentation_mymethod(train_data_matrix, train_labels_matrix, X_te, task='tissue')

err_my = util.classification_error(y_te, pred_my)
dice_my = util.dice_multiclass(y_te.reshape(-1,1), pred_my.reshape(-1,1))

print('My method on tissue task')
print('Error:', err_my)
print('Dice:', dice_my)

show_segmentation(y_te, pred_my, title='Used method: Random Forest')
```



## Research question

Does adding edge-based features improve tissue segmentation compared with using intensity and smoothing features only?

## Hypothesis

I expect edge features to improve segmentation performance, because tissue boundaries between white matter, gray matter and CSF often correspond to local intensity changes. A Sobel edge feature captures this boundary information, while raw intensity only describes the brightness of a pixel.

## Study design

I used the brain image dataset and created a pixelwise tissue classification dataset. Each pixel was one sample. The labels were background, white matter, gray matter and CSF. I trained on one subject and tested on a different subject to evaluate generalization. The features included T1 intensity, T2 intensity, Gaussian-smoothed intensities, the difference between T1 and T2, and Sobel edge strength.

## Methods

I compared different feature sets and classifiers. The feature sets were intensity only, intensity with Gaussian smoothing, intensity with Gaussian smoothing and T1-T2 difference, and all features including edge features. The classifiers were kNN and Random Forest. Performance was evaluated using classification error and multiclass Dice overlap.

## Discussion

If the Dice score improved after adding edge features, this suggests that boundary information helps the classifier distinguish tissues. If the Dice score did not improve, this suggests that edge information was not the limiting factor, or that the tissue classes overlap too much in feature space. Visual inspection of the predicted masks is important because the Dice score alone does not show where the mistakes occur.
